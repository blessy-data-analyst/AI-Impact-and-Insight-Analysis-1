#### **Chart Requirements**

#### **Top AI Tools Used**

```python
tool_usage = df['Top AI Tools Used'].value_counts()
plt.figure(figsize=(8,6))
bars = plt.bar(tool_usage.index, tool_usage.values)
plt.xticks(rotation=45)
plt.xlabel("AI Tools Used")
plt.ylabel("Count")
plt.title("Top AI Tools Used")

for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)

plt.tight_layout()
plt.show()
```

#### **AI Adoption Rate by Country**

```python
country_ai = df.groupby("Country")["AI Adoption Rate (%)"].mean().sort_values(ascending=False)
plt.figure(figsize=(8,6))
bars =plt.bar(country_ai.index, country_ai.values)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("AI Adoption Rate (%)")
plt.title("AI Adoption Rate by Country")
          
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center' , va='bottom', fontsize=8)

plt.tight_layout()
plt.show()


```

#### **AI Adoption Rate by Industry**

```python
industry_ai = df.groupby("Industry")["AI Adoption Rate (%)"].sum()
plt.figure(figsize=(4,4))

plt.pie(industry_ai , labels=industry_ai.index, 
                      autopct='%1.1f%%',
                      startangle=90)
plt.title("AI Adoption Rate by Industy")
plt.axis('equal')
plt.show()

```

#### **AI Adoption Vs Revenue**

```python
revenue_ai = df.groupby("Revenue Increase Due to AI (%)")["AI Adoption Rate (%)"].sum()
plt.figure(figsize=(4,4))

plt.scatter(df["Revenue Increase Due to AI (%)"],df["AI Adoption Rate (%)"])
plt.title("AI Adoption Vs Revenue")
plt.xlabel("AI Adoption Rate (%)")
plt.ylabel("Revenue Index")
plt.show()
```

#### **AI Adoption Vs Job Loss by Country**

```python
country_data = df.groupby("Country").mean(numeric_only=True)
fig, ax1 = plt.subplots(figsize=(8,4))
ax1.bar(country_data.index , country_data["AI Adoption Rate (%)"],color="green")
ax1.set_ylabel("AI Adoption Rate (%)")
plt.xticks(rotation=45)

ax2= ax1.twinx()
ax2.plot(country_data.index, country_data["Job Loss Due to AI (%)"],color="Orange" ,marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel("Job Loss Due to AI (%)")


plt.title("AI Adoption Vs Job Loss by Country")
plt.show()
```

#### **AI Adoption Trend over Years**

```python
year_ai = df.groupby("Year")["AI Adoption Rate (%)"].mean()
plt.figure(figsize=(8,4))
plt.plot(year_ai.index, year_ai.values, marker='o',linestyle='-')
plt.title("AI Adoption Trend")
plt.xlabel("year")
plt.ylabel("AI Adoption Rate (%)")

for x,y in zip(year_ai.index, year_ai.values):
     plt.text(x,y, f'{y:,.0f}',ha='center',va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Revenue Growth by Industry**

```python
revenue_industry = df.groupby("Industry")["Revenue Increase Due to AI (%)"].sum()
plt.figure(figsize=(8,6))
bars=plt.bar(revenue_industry.index, revenue_industry.values, color='Brown')
plt.xticks(rotation=-90)
plt.xlabel("Industry")
plt.ylabel("Revenue Increase Due to AI (%)")
plt.title=("Revenue Growth by Industry")

for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center',va='bottom',fontsize=8,)

plt.tight_layout()
plt.show()
```

#### **Market Share by Regulation Status**

```python

plt.figure(figsize=(10,6))

df.boxplot(column="Market Share of AI Companies (%)", by="Regulation Status")
plt.xlabel("Regulation Status")
plt.ylabel("Market Share")

plt.show()
```

#### **HumanAI Collaboration Rate Distribution by Country**

```python
collab_country = df.groupby("Country")["Human-AI Collaboration Rate (%)"].sum()

plt.figure(figsize=(8,5))
plt.pie(collab_country, labels=collab_country.index, autopct='%1.1f%%')
plt.title("Human-AI Collaboration by Country")
plt.show()
```

#### **AI-Generated Content by Industry**

```python
industry_content = df.groupby("Industry")["AI-Generated Content Volume (TBs per year)"].mean()

fig,ax=plt.subplots(figsize=(6,5))
industry_content.plot(kind='bar',ax=ax)
plt.xticks(rotation=45)
ax.set_title("AI-Generated Content by Industry")
ax.set_xlabel("Industry")
ax.set_ylabel("Content Volume (TBs/year)")

for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width()/2,bar.get_height(),
             f'{bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)
    
plt.tight_layout()
plt.show()
```
